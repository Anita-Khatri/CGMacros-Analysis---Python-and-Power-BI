```python
#for subject 1 we are adding 'subject' number and extracting 'photo path' and 'formatted datatime' 
# and then will match the formatted datetime and subject with the cosolidated file.

import pandas as pd
import re
from datetime import datetime

# Load the CSV
df = pd.read_csv('Subject_1_photo_files_names.csv')

# Add the 'photos/' prefix
df['Photo Path'] = 'photos/' + df['filename']

# Function to extract datetime and format as 'yyyy/mm/dd hh:00:00'
def extract_datetime(filename):
    match = re.search(r'(\d{4})-(\d{1,2})-(\d{1,2})-(\d{1,2})-(\d{1,2})-\d+', filename)
    if match:
        year, month, day, hour, minute = map(int, match.groups())
        try:
            dt = datetime(year, month, day, hour, minute)
            return dt.strftime('%Y/%m/%d %H:00:00')
        except ValueError:
            return None
    return None

# Extract datetime
df['Formatted Datetime'] = df['filename'].apply(extract_datetime)

# Add a new column 'subject' with constant value 1
df['subject'] = 1

# Save to CSV
df.to_csv('Subject1_photo_names_with_datetime_and_Subject.csv', index=False)
```

```python
#Subject 1 and Formated datetime are matched with "suject" and "start_hour" of "subjectsHourlyAveragedCGMacros"

import pandas as pd
from datetime import datetime

# Load CSVs
photos_df = pd.read_csv('Subject1_photo_names_with_datetime_and_Subject.csv')
hourly_df = pd.read_csv('SubjectsHourlyAveragedCGMacros.csv')

# Convert datetime columns to datetime objects
photos_df['Formatted Datetime'] = pd.to_datetime(photos_df['Formatted Datetime'], format='%Y/%m/%d %H:%M:%S')
hourly_df['hour_start'] = pd.to_datetime(hourly_df['hour_start'])

# Ensure only the first photo per subject per hour is used
photos_df = photos_df.sort_values(['subject', 'filename'])  # Optional sort
photos_df = photos_df.drop_duplicates(subset=['subject', 'Formatted Datetime'], keep='first')

# Merge on subject and datetime
merged_df = pd.merge(hourly_df, photos_df, how='left',
                     left_on=['subject', 'hour_start'],
                     right_on=['subject', 'Formatted Datetime'])

# Update 'Image path' only if it's missing and a photo path is available
merged_df['Image path'] = merged_df.apply(
    lambda row: row['Photo Path'] if (pd.isna(row['Image path']) or row['Image path'] == '') and pd.notna(row['Photo Path']) else row['Image path'],
    axis=1
)

# Keep only original columns (with updated 'Image path')
final_df = merged_df[hourly_df.columns]

# Save the result
final_df.to_csv('SubjectHourlyAveragedCGMacros_UpdatedPhotoPaths.csv', index=False)
```
