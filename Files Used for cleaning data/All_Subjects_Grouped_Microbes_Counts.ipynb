```python
import pandas as pd

# Step 1: Load CSV
df = pd.read_csv("microbes.csv")


# Step 2: Keyword-to-group mapping (all lowercase keys for consistency)
keyword_to_group = {
    'lactobacillus': "Potential Probiotic/Bioremediation",
    'limosilactobacillus': 'Potential Probiotic/Bioremediation',
    'loigolactobacillus': 'Potential Probiotic/Bioremediation',
    'bifidobacterium': 'Potential Probiotic/Bioremediation',
    'bacillus': 'Industrial',
    'megamonas': 'Amino Acid Fermenters',
    'megasphaera': 'Amino Acid Fermenters',
    'methano': 'Methanogens',
    'methanobrevibacter': 'Methanogens',
    'methanosphaera': 'Methanogens',
    'mucispirillum': 'Mucin Degraders',
    'malassezia': 'Oral Microbiota',
    'moraxella': 'Pathogens',
    'mobiluncus': 'Opportunistic Pathogens',
    'mogibacterium': 'Opportunistic Pathogens',
    'mitsuokella': 'Lactic Acid Producers',
    'parabacteroides': 'Gut Microbiota',
    'oscillibacter': 'Fiber Degraders',
    'lysobacter': 'Environmental',
    'mesorhizobium': 'Environmental',
    'microbacterium': 'Environmental',
    'marvinbryantia': 'Butyrate Producers',
    'mediterraneibacter': 'Butyrate Producers',
    'virus': 'Plant Virus',
    'candida': 'Pathogens',
    'aspergillus': 'Pathogens',
    'penicillium': 'Pathogens',
    'pseudomonas': 'Pathogens',
    'escherichia': 'Pathogens',
    'salmonella': 'Pathogens',
    'shigella': 'Pathogens',
    'treponema': 'Pathogens',
    'fusobacterium': 'Pathogens',
    'prevotella': 'Oral Microbiota',
    'porphyromonas': 'Pathogens',
    'neisseria': 'Oral Microbiota',
    'mycobacterium': 'Obligate Parasite with Symbiotic Bacteria',
    'saccharomyces': 'Industrial',
    'faecalibacterium': 'Gut Microbiota',
    'roseburia': 'Butyrate Producers',
    'escherichia coli': 'Model Organisms',
    'bacteroides': 'Carbohydrate Fermenters',
    'ruminococcus': 'Carbohydrate Fermenters',
    'akkermansia': 'Mucin Degraders',
    'blautia': 'SCFA Producers',
    'clostridium': 'Amino Acid Fermenters',
    'alistipes': 'Amino Acid Fermenters',
    'bilophila': 'Sulfate Reducers',
    'streptococcus': 'Lactic Acid Producers',
    'enterococcus': 'Lactic Acid Producers',
    'abiotrophia': 'Opportunistic Pathogens',
    'acetivibrio': 'Fiber Degraders',
    'achromobacter': 'Environmental',
    'acidaminococcus': 'Amino Acid Fermenters',
    'acidovorax': 'Opportunistic Pathogens',
    'acinetobacter': 'Opportunistic Pathogens',
    'actinobaculum': 'Opportunistic Pathogens',
    'actinomyces': 'Oral Microbiota',
    'collinsella': 'Carbohydrate Fermenters',
    'dialister': 'Opportunistic Pathogens',
    'eubacterium': 'Butyrate Producers',
    'propionibacterium': 'Skin/Oral Microbiota',
    'adlercreutzia': 'Gut Microbiota',
    'aerococcus': 'Opportunistic Pathogens',
    'aeromonas': 'Opportunistic Pathogens',
    'afipia': 'Opportunistic Pathogens',
    'agrobacterium': 'Environmental',
    'aggregatibacter': 'Pathogens',
    'alloprevotella': 'Oral Microbiota',
    'alphaendornavirus': 'Plant Virus',
    'alternaria': 'Pathogens',
    'amedibacterium': 'Gut Microbiota',
    'aminipila': 'Carbohydrate Fermenters',
    'anaerobutyricum': 'Butyrate Producers',
    'anaerococcus': 'Oral Microbiota',
    'anaerofustis': 'SCFA Producers',
    'anaeroglobus': 'SCFA Producers',
    'anaerosphaera': 'Amino Acid Fermenters',
    'anaerosporobacter': 'Amino Acid Fermenters',
    'anaerostipes': 'Butyrate Producers',
    'anaerotruncus': 'SCFA Producers',
    'angelakisella': 'Gut Microbiota',
    'angomonas': 'Parasite',
    'anoxybacillus': 'Environmental',
    'apple': 'Plant Virus',
    'aquabacterium': 'Environmental',
    'arachnia': 'Oral Microbiota',
    'atlantibacter': 'Opportunistic Pathogens',
    'atopobium': 'Oral/Genital Microbiota',
    'aureobasidium': 'Opportunistic Pathogens',
    'avibacterium': 'Animal Pathogen',
    'bacteroidales': 'Gut Microbiota',
    'bariatricus': 'Unknown',
    'barnesiella': 'Gut Microbiota',
    'berryella': 'Unknown',
    "beet necrotic yellow virus": "Plant Virus",
    "bittarella": "Unknown",
    "blastocystis": "Parasite",
    "blumeria": "Plant Pathogen",
    "bordetella": "Opportunistic Pathogens",
    "borreliella": "Tick-borne Pathogen",
    "botrytis": "Pathogens",
    "bradyrhizobium": "Environmental",
    "brochothrix": "Environmental",
    "brome": "Unknown",
    "bubaline": "Unknown",
    "bulleidia":"Opportunistic Pathogens",
     "burkholderiales": "Opportunistic Pathogens",
    "butyricicoccus": "Butyrate Producers",
    "butyricimonas": "SCFA Producers",
    "butyrivibrio": "Butyrate Producers",
    "cactus": "Plant Virus",
    "caenorhabditis": "Model Organism",
    "capnocytophaga": "Opportunistic Pathogens",
    "cardiobacterium": "Pathogens",
    "carnobacterium": "Environmental",
    "catabacter": "Gut Microbiota",
    "catenibacterium": "Carbohydrate Fermenters",
    "catonella": "Oral Microbiota",
    "caulobacter": "Environmental",
    "cellulomonas": "Environmental",
    "christensenella": "Gut Microbiota",
    "chryseobacterium": "Opportunistic Pathogens",
    "citrobacter": "Opportunistic Pathogens",
     "clavispora": "Opportunistic Pathogens",
    "cloacibacillus": "Amino Acid Fermenters",
    "clostridia": "Amino Acid Fermenters",
    "clostridiaceae": "SCFA Producers",
    "clostridiales": "Amino Acid Fermenters",
    "clostridioides": "Opportunistic Pathogens",
    "comamonas": "Opportunistic Pathogens",
    "coprobacillus": "Butyrate Producers",
    "coprobacter": "Gut Microbiota",
    "coprococcus": "Gut Microbiota",
    "corynebacterium": "Skin/Oral Microbiota",
    "criibacterium": "Unknown",
    "cronobacter": "Opportunistic Pathogens",
    "cryptobacterium": "Oral Microbiota",
    "cucumber": "Plant Virus",
    "cucurbit": "Virus",
    "culturomica": "Environmental",
    "cupriavidus": "Environmental",
    "cutibacterium": "Skin/Oral Microbiota",
    "debaryomyces": "Opportunistic Pathogens",
    "delftia": "Opportunistic Pathogens",
    "desulfobulbus": "Sulfate Reducers",
    "desulfovibrio": "Sulfate Reducers",
     "dielma": "Gut Microbiota",
    "dietzia": "Opportunistic Pathogens",
    "dorea": "Gut Microbiota",
    "drancourtella": "Unknown",
    "duncaniella": "Gut Microbiota",
    "dysosmobacter": "Gut Microbiota",
    "eggerthella": "Opportunistic Pathogens",
    "eisenbergiella": "Gut Microbiota",
    "elizabethkingia": "Opportunistic Pathogens",
    "emergencia": "Unknown",
    "entamoeba": "Parasite",
    "enterobacter": "Opportunistic Pathogens",
    "enterocloster": "Gut Microbiota",
    "enterorhabdus": "Gut Microbiota",
    "eptesicus": "Virus",
    "erysipelatoclostridium": "SCFA Producers",
    "erysipelothrix": "Animal Pathogen",
    "erysipelotrichaceae": "Gut Microbiota",
    "exiguobacterium": "Environmental",
    "exophiala": "Pathogens",
    "ezakiella": "Oral/Genital Microbiota",
    "facklamia": "Opportunistic Pathogens",
    "faecalibacillus": "Gut Microbiota",
    "faecalicoccus": "Gut Microbiota",
    "faecalitalea": "Gut Microbiota",
    "fannyhessea": "Oral/Genital Microbiota",
    "fastidiosipila": "Oral/Gut Microbiota",
    "fenollaria": "Gut Microbiota",
    "fibrobacter": "Cellulose Degrader",
    "filifactor": "Pathogens",
    "finegoldia": "Opportunistic Pathogens",
    "firmicutes": " Mixed Functions",
    "flavobacterium": "Environmental",
    "flavonifractor": "Butyrate Producers",
     "flintibacter": "Gut Microbiota",
    "fournierella": "Gut Microbiota",
    "furfurilactobacillus": "Potential Probiotic",
    "fusicatenibacter": "Butyrate Producers",
    "gabonia": "Unknown",
    "gardnerella": "Opportunistic Pathogens",
    "garlic": "Plant Virus",
    "gemella": "Oral/ Respiratory Microbiota",
    "geobacillus": "Environmental",
    "geobacter": "Environmental",
    "glaesserella": "Animal Pathogen",
    "gordonibacter": "Gut Microbiota",
    "gorganvirus": "Virus",
    "granulicatella": "Opportunistic Pathogens",
    "grapevine": "Unknown",
    "haemophilus": "Opportunistic Pathogens",
    "hafnia": "Opportunistic Pathogens",
    "hallella": "Oral/Gut Microbiota",
    "helianthus": "Unknown",
    "helicobacter": "Pathogens",
    "hericium": "Fungus",
    "hespellia": "Gut Microbiota",
    "holdemanella": "Gut Microbiota",
    "holdemania": "SCFA Producers",
    "human": "Virus",
    "hungatella": "Pathogens",
    "hyphopichia": "Fungus",
    "idiomarina": "Environmental",
    "intestinibacter": "Gut Microbiota",
    "intestinibaculum": "Gut Microbiota",
    "intestinimonas": "Gut Microbiota",
    "johnsonella": "Oral Microbiota",
    "kallipyga": "Oral Microbiota",
    "kandleria": "Oral Microbiota",
    "kingella": "Opportunistic Pathogens",
    "klebsiella": "Opportunistic Pathogens",
    "kluyvera": "Opportunistic Pathogens",
    "kocuria": "Environmental",
    "lachancea": "Environmental",
    "lachnoanaerobaculum": "Oral Microbiota",
    "lachnoclostridium": "Gut Microbiota",
    "lachnospira": "Gut Microbiota",
    "lachnospiraceae": "SCFA Producers",
    "lacrimispora": "Butyrate Producers",
    "lacticaseibacillus": "Potential Probiotic",
    "lactiplantibacillus": "Potential Probiotic",
    "lactococcus": "Lactic Acid Producers",
    "lactonifactor": "Gut Microbiota",
    "lancefieldella": "Gut Microbiota",
    "latilactobacillus": "Lactic Acid Producers",
    "lautropia": "Opportunistic Pathogens",
    "leclercia": "Opportunistic Pathogens",
    "lelliottia": "Opportunistic Pathogens",
    "leptotrichia": "Opportunistic Pathogens",
    "leuconostoc": "Lactic Acid Producers",
    "levilactobacillus": "Lactic Acid Producers",
    "ligilactobacillus": "Lactic Acid Producers",
    "levyella": "Unknown",
    "longibaculum": "Unknown",
    "macrococcus": "Opportunistic Pathogens",
    "mageeibacillus": "Vaginal Microbiota",
    "mannheimia": "Animal Pathogen",
    "massilioclostridium": "Gut Microbiota",
    "massiliomicrobiota": "Unknown",
    "massilistercora": "Unknown",
    "mediterranea": "Environmental",
    "merdibacter": "Gut Microbiota",
    "morganella": "Opportunistic Pathogens",
    "murdochiella": "Gut Microbiota",
    "muribaculum": "SCFA Producers",
    "mycoplasma": "Parasite",
    "myroides": "Opportunistic Pathogens",
    "nannochloropsis": "Algae",
    "ndongobacter": "Unknown",
    "nectarine": "Unknown",
    "nectria": "Plant Pathogen",
    "negativicoccus": "Gut Microbiota",
    "neglecta": "Unknown",
    "obesumbacterium": "Opportunistic Pathogens",
    "ochrobactrum": "Opportunistic Pathogens",
    "odoribacter": "SCFA Producers",
    "olsenella": "Oral/Gut Microbiota",
    "oribacterium": "Oral Microbiota",
    "ornithobacterium": "Avian Pathogen",
    "oscillospiraceae": "SCFA Producers",
    "oxalobacter": "Gut Microbiota",
    "paenarthrobacter": "Environmental",
    "paenibacillus": "Nitrogen Fixation",
    "paeniclostridium": "SCFA Producers",
    "pannonibacter": "Environmental",
    "pantoea": "Opportunistic Pathogens",
    "paracoccus": "Environmental",
    "paraprevotella": "Gut Microbiota",
    "parasutterella": "Gut Microbiota",
    "parolsenella": "Gut Microbiota",
    "parvimonas": "Opportunistic Pathogens",
    "pediococcus": "Lactic Acid Producers",
    "peptacetobacter": "Gut Microbiota",
    "peptoniphilus": "Opportunistic Pathogens",
    "peptostreptococcus": "Opportunistic Pathogens",
    "petrimonas": "Gut Microbiota",
    "phascolarctobacterium": "SCFA Producers",
    "phocaeicola": "Gut Microbiota",
    "plantactinospora": "Environmental",
    "polerovirus": "Plant Virus",
    "pontibacillus": "Environmental",
    "priestia": "Environmental",
    "propionimicrobium": "Skin/Oral Microbiota",
    "proteus": "Opportunistic Pathogens",
    "providencia": "Opportunistic Pathogens",
    "pseudobutyrivibrio": "SCFA Producers",
    "pseudoflavonifractor": "SCFA Producers",
    "pseudoramibacter": "SCFA Producers",
    "ralstonia": "Plant Pathogen",
    "raoultella": "Opportunistic Pathogens",
    "riemerella": "Avian Pathogen",
    "rikenella": "Gut Microbiota",
    "romboutsia": "SCFA Producers",
    "rothia": "Opportunistic Pathogens",
    "ruminiclostridium": "Fiber Degraders",
    "ruminococcaceae": "Fiber Degraders",
    "ruthenibacterium": "Gut Microbiota",
    "sanguibacteroides": "Gut Microbiota",
    "sarcina": "Environmental",
    "scardovia": "Oral Microbiota",
    "schaalia": "Oral Microbiota",
    "schlumbergera": "Unknown",
    "selenomonas": "Oral/Gut Microbiota",
    "sellimonas": "SCFA Producers",
    "senegalimassilia": "Gut Microbiota",
    "serratia": "Opportunistic Pathogens",
    "sfi11virus": "Bacteriophage",
    "sfi21dt1virus": "Bacteriophage",
    "sharpea": "Gut Microbiota",
    "shewanella": "Opportunistic Pathogens",
    "shuttleworthia": "Gut Microbiota",
    "sinorhizobium": "Environmental",
    "slackia": "Oral/Gut Microbiota",
    "sneathia": "Oral/Genital Microbiota",
    "sodaliphilus": "Environmental",
    "solobacterium": "Oral Microbiota",
    "sparassis": "Fungus",
    "sphingobacterium": "Environmental",
    "sphingomonas": "Environmental",
    "sphingopyxis": "Environmental",
    "staphylococcus": "Opportunistic Pathogens",
    "stenotrophomonas": "Opportunistic Pathogens",
    "stomatobaculum": "Oral Microbiota",
    "streptomyces": "Antibiotic-Producing Actinobacteria",
    "subdoligranulum": "SCFA Producers",
    "sutterella": "Gut Microbiota",
    "sutterellaceae": "Gut Microbiota",
    "synergistes": "Gut Microbiota",
    "talaromyces": "Opportunistic Pathogens",
    "tannerella": "Pathogens",
    "tepidimonas": "Environmental",
    "terrisporobacter": "SCFA Producers",
    "tidjanibacter": "Rare",
    "tissierellia": "Gut Microbiota",
    "tobacco": "Plant Virus",
    "tobamovirus": "Plant Virus",
    "tomato": "Plant Virus",
    "toxoplasma": "Parasite",
    "traorella": "Rare",
    "tropical": "Plant Virus",
    "trypanosoma": "Parasite",
    "turicibacter": "Gut Microbiota",
    "tyzzerella": "Gut Microbiota",
    "ureaplasma": "Pathogens",
    "urmitella": "Unknown",
    "valsa": "Plant Pathogen",
    "variovorax": "Environmental",
    "watermelon": "Plant Virus",
    "weissella": "Lactic Acid Producers",
    "weizmannia": "Industrial",
    "white": "Virus",
    "zygosaccharomyces": "Fungus"
    
    
}

# Step 3: Clean column names (remove leading/trailing spaces)
df.columns = df.columns.str.strip()


# Remove square brackets from values in the 'column_name' column
df.columns = [str(col) for col in df.columns]
df.columns = pd.Index(df.columns).str.replace(r'\[|\]', '', regex=True).str.strip()


# Step 4: Extract bacteria columns (excluding 'subject')
bacteria_cols = [col for col in df.columns if col.lower() != 'subject']


# Step 5: Create mappings
genus_mapping = {}
genus_to_group = {}

for col in bacteria_cols:
    genus = col.split()[0].lower()
    genus_mapping[col] = genus

    group = keyword_to_group.get(genus)
    if group:
        genus_to_group[col] = group
    else:
        for keyword, value in keyword_to_group.items():
            if genus.startswith(keyword):
                genus_to_group[col] = value
                break
        else:
            genus_to_group[col] = 'Unknown'


# Step 6: Transpose and assign Genus + Group
df_transposed = df[bacteria_cols].T.copy()
df_transposed['Group'] = df_transposed.index.map(genus_to_group)
df_transposed['Genus'] = df_transposed.index.map(genus_mapping)


# Ensure only numeric data is used for summing
df_transposed_filtered = df_transposed.drop(columns=['Group', 'Genus'])
df_transposed_filtered = df_transposed_filtered.apply(pd.to_numeric, errors='coerce').fillna(0)


# Add Group and Genus back
df_transposed_filtered['Group'] = df_transposed['Group']
df_transposed_filtered['Genus'] = df_transposed['Genus']


# Keep only rows that are assigned to a known genus (real microbes)
valid_rows = df_transposed_filtered['Genus'].notna()
df_transposed_filtered = df_transposed_filtered[valid_rows]


# Step 7: Group by functional category
grouped = df_transposed_filtered.groupby('Group').sum().T

# Step 8: Insert subject column back if present
if 'subject' in df.columns:
    grouped.insert(0, 'subject', df['subject'])

#grouped_cleaned = grouped[pd.to_numeric(grouped['subject'], errors='coerce').notna()]
grouped = grouped[pd.to_numeric(grouped['subject'], errors='coerce').notna()]


# Step 9: Save outputs
grouped.to_csv("grouped_microbes_final.csv", index=False)
df_transposed_filtered.to_csv("microbe_genus_group_mapping.csv", index=False)


print("✅ Saved:")
print("- Functional group counts per subject → grouped_microbes_final.csv")
print("- Genus + Group mapping per microbe → microbe_genus_group_mapping.csv")

```
